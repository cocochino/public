## Poppular Dog Breeds

### What is the most popular breed?

My SQL doesn't have TOP function. TO select top item, use ORDER BY... LIMIT 1
~~~
SELECT breed, count(breed) from ny_dogs
GROUP BY 1 ORDER BY 2 DESC
LIMIT 1;
~~~
Breed | Count 
----| ----
Mixed/Other | 23170

No surprise here.

### Mixed/Others in each borough
~~~
SELECT borough, count(breed) from ny_dogs
where breed = 'Mixed/Other'
GROUP BY 1 ORDER BY 2 DESC
